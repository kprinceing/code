### 青年

```python
input_files = list(pathlib.Path('/Users/yansong/Nutstore Files/基金研究/data/2013--2022面青/scopus').glob('*.csv'))

doc_cols = ['author_eid','eid','doi','title','afid','affilname','author_names','coverDate','citedby_count']
doc_cols2 = ['eid','doi','title','afid','affilname','author_names','coverDate','citedby_count','publicationName']
counter = 0
for i in input_files:
    applicant_df = pd.read_csv(i)
    applicant_eid = list(applicant_df['eid'])
    applicant_eid_df = pd.DataFrame(columns=doc_cols)
    i_str = str(i)
    output_file_name = re.sub("scopus", "scopusPub_New", i_str)
    if os.path.exists(output_file_name):
        counter = counter+1
    else:
        display('new')
        for j in applicant_eid:
            eid_num = j[7:]
            au = AuthorRetrieval(eid_num)
            docs = pd.DataFrame(au.get_documents(refresh=10))
            docs_output = docs[doc_cols2]
            docs_output['author_eid'] = eid_num
            applicant_eid_df= pd.concat([applicant_eid_df,docs_output])
        filepath = Path(output_file_name)  
        filepath.parent.mkdir(parents=True, exist_ok=True)
        applicant_eid_df.to_csv(filepath)  
```
